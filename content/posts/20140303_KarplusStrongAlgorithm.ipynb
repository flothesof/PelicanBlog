```python
import StringIO
import base64
import struct
from IPython.core.display import HTML

def wavPlayer(data, rate):
    """ will display html 5 player for compatible browser
    The browser need to know how to play wav through html5.
    there is no autoplay to prevent file playing when the browser opens
    Adapted from SciPy.io.
    """
    
    buffer = StringIO.StringIO()
    buffer.write(b'RIFF')
    buffer.write(b'\x00\x00\x00\x00')
    buffer.write(b'WAVE')

    buffer.write(b'fmt ')
    if data.ndim == 1:
        noc = 1
    else:
        noc = data.shape[1]
    bits = data.dtype.itemsize * 8
    sbytes = rate*(bits // 8)*noc
    ba = noc * (bits // 8)
    buffer.write(struct.pack('<ihHIIHH', 16, 1, noc, rate, sbytes, ba, bits))

    # data chunk
    buffer.write(b'data')
    buffer.write(struct.pack('<i', data.nbytes))

    if data.dtype.byteorder == '>' or (data.dtype.byteorder == '=' and sys.byteorder == 'big'):
        data = data.byteswap()

    buffer.write(data.tostring())
    # Determine file size and place it in correct
    #  position at start of the file.
    size = buffer.tell()
    buffer.seek(4)
    buffer.write(struct.pack('<i', size-8))
    
    val = buffer.getvalue()
    
    src = """
    <head>
    <meta http-equiv="Content-Type" content="text/html; charset=utf-8">
    <title>Simple Test</title>
    </head>
    
    <body>
    <audio controls="controls" style="width:600px" >
      <source controls src="data:audio/wav;base64,{base64}" type="audio/wav" />
      Your browser does not support the audio element.
    </audio>
    </body>
    """.format(base64=base64.encodestring(val))
    display(HTML(src))
```

First let's start with the direct implementation of the Karplus Strong algorithm.

```python
%pylab inline
```

```python
def ks_loop(x, alpha, D):
    M = len(x)
    size_y = D * M
    y = zeros((size_y))
    y[:M] = x[:]
    for index in range(M, size_y):
        y[index] = alpha * y[index - M]
    return y
```

```python
x = randn(200)
```

```python
plot(x, '-o')
```

```python
%%time
y = ks_loop(x, 0.9, 50)
```

```python
plot(y, '-o')
```

Let's listen to that!

```python
wavPlayer(2**13 * y.astype(int16), 8000)
```

Now, let's move on to the vectorized implementation.

```python
def ks(x, alpha, D):
    return concatenate([x * alpha ** i for i in range(D)])
```

```python
%%time
ks(x, 0.9, 10)
```

```python
allclose(ks(x, 0.9, 10), ks_loop(x, 0.9, 10))
```

As the last output shows, we have built a function that yields the same previous result than our previous looping function, only faster.

```python
%timeit ks(x, 0.9, 10)
%timeit ks_loop(x, 0.9, 10)
```

Now, let's move on to the last step : **A Hard Day's Night** by the *Beatles*.

```python
def frequency(semitones):
    return 440 * 2**(semitones/12.)
```

```python
frequency(12)
```

```python
Fs = 48000 # sampling frequency
notes = [-31, -19, -16, -14, -4, 0, 3, 10] # D2, D3, F3, G3, F4, A4, C5, G5
gains = [1.2, 3.0, 1.0, 2.2, 1.0, 1.0, 1.0, 3.5]
duration = 4
alpha = 0.9875
nbsample_chord = Fs * duration # number of samples in sound
first_duration = nbsample_chord / Fs * frequency(notes[0])

chord = zeros(nbsample_chord)
for index, note in enumerate(notes):
    current_M = ceil(Fs/frequency(note)).astype(int)
    current_duration = ceil(nbsample_chord / current_M).astype(int) + 1
    current_alpha = alpha ** (first_duration / current_duration)
    if index == 1:
        current_alpha = current_alpha ** .8
    x = randn(current_M)
    y = ks(x, current_alpha, current_duration)
    chord += gains[index] * y[:nbsample_chord]
```

```python
plot(chord)
```

```python
wavPlayer(2**13 * chord.astype(int16), Fs)
```

Now that we have done this sort of analysis, I would like to finish with a little study inspired by the [Wikipedia page for the Karplus Strong algorithm](http://en.wikipedia.org/wiki/Karplus%E2%80%93Strong_string_synthesis) that exhibits a [nice Vorbis file simulating different decays](http://en.wikipedia.org/wiki/File:Karplus-strong-A2.ogg). As the description tells us, we will model a sound with a frequency of 220 Hz at different decay rates.

```python
Fs = 48000 # sampling rate
f_0 = 220 # the frequency of the sound we will play
duration = 1 # duration of the sound
nbsample = duration * Fs # number of samples at given sampling frequency
x = randn(int(Fs/f_0))
play_sound = lambda attenuation : wavPlayer(2**13 * ks(x, attenuation, nbsample / x.size).astype(int16), Fs)
```

```python
play_sound(0.97)
```

```python
for att in [0.995, 0.99, 0.985, 0.98, 0.975, 0.5]:
    play_sound(att)
```

Comparing with the original, it seems that the sound does not sound like a guitar sound due to the filtering which is missing in this version. To achieve a better sound, it would be necessary to use a low-pass filter on top of the KS algorithm as used here.

One possibility of how to do this is to use a partial an averaging method, therefore implementing this sort of filter.

```python
def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n

def ks_filtered(x, alpha, D, averaging):
    y = concatenate([moving_average(x, averaging) * alpha ** i for i in range(D)])
    return  y / abs(y).max() * 4.
```

```python
plot(ks(x, 0.9, 10), label='original KS')
plot(ks_filtered(x, 0.90, 10, 5), label='filtered KS')
legend()
```

```python
%timeit ks(x, 0.9, 10)
%timeit ks_filtered(x, 0.9, 10, 3)
```

```python
play_filtered_sound = lambda averaging : wavPlayer(2**13 * ks_filtered(x, 0.995, nbsample / x.size, averaging).astype(int16), Fs)
```

```python
play_filtered_sound(5)
```

```python
for averaging in range(1, 100, 10):
    play_filtered_sound(averaging)
```

As you can see, the filtering alters the sound, but also the mean frequency, which goes lower and lower.

```python
import collections
def ks_stanford(x, alpha, D):
    d = collections.deque(x, x.size)
    output = zeros((D * x.size))
    output[:x.size] = x
    for i in range(D):
        for j in range(x.size):
            d.append(alpha / 2. * (d[0] + d[1]))
        output[i * x.size: (i+1)*x.size] = array(d)
    return output
```

```python
plot(ks_stanford(x, 1., 100))
```

```python
fs = 44000
f0 = 880
n = fs / f0
dt = 4.
N = dt * fs
x = rand(n)
wavPlayer(2**13 * ks_stanford(randn(n), 0.999999, int(N/f0)).astype(int16), fs)
```
